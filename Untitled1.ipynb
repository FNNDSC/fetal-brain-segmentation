```python
import os
import sys
import random
import math
import re
import time
import numpy as np
import tensorflow as tf
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Root directory of the project
ROOT_DIR = os.path.abspath("../../")

# Import Mask RCNN
sys.path.append(ROOT_DIR)  # To find local version of the library
from mrcnn import utils
from mrcnn import visualize
from mrcnn.visualize import display_images
import mrcnn.model as modellib
from mrcnn.model import log

from cortical_seg import *

%matplotlib inline 

# Directory to save logs and trained model
LOGS_DIR = os.path.join(ROOT_DIR, "logs")

```

```python
def get_ax(rows=1, cols=1, size=16):
    """Return a Matplotlib Axes array to be used in
    all visualizations in the notebook. Provide a
    central point to control graph sizes.
    
    Adjust the size attribute to control how big to render images
    """
    fig, ax = plt.subplots(rows, cols, figsize=(size*cols, size*rows))
    fig.tight_layout()
    return ax
```

```python
import glob

patients_list = [x[x.find('/') + 1:x.rfind('/')] for x in glob.glob('data/*/*')]

random.seed(4)
random.shuffle(patients_list)

patients_list = set(patients_list)

data = []
for patient in patients_list:
    img = glob.glob('data/%s/recon*'%patient)[0]
    mask = glob.glob('data/%s/seg*'%patient)[0]
    data.append((img,mask))

split = int(len(data) * 0.8)
#TODO CAMBIA ESTO
val_data = data[:split]

val_dataset = CorticalSegDataset()
val_dataset.load_data(val_data)
val_dataset.prepare()
```

```python
model = modellib.MaskRCNN(mode="inference",
                              model_dir=LOGS_DIR,
                              config=config)
```

```python
# Or, load the last model you trained
weights_path = model.find_last()

# Load weights
print("Loading weights ", weights_path)
model.load_weights(weights_path, by_name=True)
```

```python
import skimage.color

image_id = random.choice(val_dataset.image_ids)
print(image_id)

image, image_meta, gt_class_id, gt_bbox, gt_mask =\
    modellib.load_image_gt(val_dataset, config, image_id, use_mini_mask=False)

print(image.shape)

# Run object detection
results = model.detect_molded(np.expand_dims(image, 0), np.expand_dims(image_meta, 0), verbose=1)

# Display results
r = results[0]
log("gt_class_id", gt_class_id)
log("gt_bbox", gt_bbox)
log("gt_mask", gt_mask)

# Compute AP over range 0.5 to 0.95 and print it
utils.compute_ap_range(gt_bbox, gt_class_id, gt_mask,
                       r['rois'], r['class_ids'], r['scores'], r['masks'],
                       verbose=1)

image = np.squeeze(image)
image = skimage.color.gray2rgb(image)

print(val_dataset.class_names)

visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], 
                            val_dataset.class_names, r['scores'], ax=get_ax(),
                            title="Predictions")

'''visualize.display_differences(
    image,
    gt_bbox, gt_class_id, gt_mask,
    r['rois'], r['class_ids'], r['scores'], r['masks'],
    val_dataset.class_names, ax=get_ax(),
    show_box=True, show_mask=True,
    iou_threshold=0.5, score_threshold=0.5)'''
```
